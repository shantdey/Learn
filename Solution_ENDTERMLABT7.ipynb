```python
import nltk
from nltk.corpus import wordnet as wn

# Download once: nltk.download('wordnet')

# Synonyms and hypernyms for "car"
car_synsets = wn.synsets('car')
synonyms = set(lemma.name().replace('_', ' ') for s in car_synsets for lemma in s.lemmas())
hypernyms = set(lemma.name().replace('_', ' ') for s in car_synsets[0].closure(lambda x: x.hypernyms()) for lemma in s.lemmas())
print(wn.synsets('car')[0])
print("Car synonyms:", list(synonyms))
print("Car hypernyms:", list(hypernyms))

# Path similarity between car and motorcar
sim = wn.synsets('car')[0].path_similarity(wn.synsets('motorcar')[0])
print(f"\nPath similarity (car, motorcar): {sim}")  # Should be 1.0
```

```python
import nltk
from nltk.corpus import gutenberg
from nltk import FreqDist
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures
import string

# Download if needed (once)
nltk.download('gutenberg')
nltk.download('punkt')

# Load words, clean (lowercase, alpha only)
words = gutenberg.words('shakespeare-caesar.txt')
clean_words = [w.lower() for w in words if w.isalpha()]
#top 10 word frequencies
fdist = FreqDist(clean_words)
top_words = fdist.most_common(10)
print("Top 10 words:", top_words)

#top 5 Bigrams
bigram_measures = BigramAssocMeasures()
finder = BigramCollocationFinder.from_words(clean_words)
finder.apply_freq_filter(3)  # Filter low-freq
top_bigrams = finder.nbest(bigram_measures.pmi, 5)
print("Top 5 bigrams:", top_bigrams)
```

```python
#POS Tagging
import nltk
from nltk import pos_tag, word_tokenize

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

sentence = "Chatbots are essential in the Post-Sale Customer Service Process."
tokens = word_tokenize(sentence)
tagged = pos_tag(tokens)
print(tagged)
```

```python
#Stopword filtering
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
tokens = word_tokenize("Chatbots are essential in the customer service process.".lower())
filtered = [w for w in tokens if w.isalpha() and w not in stop_words]
print(filtered)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import FreqDist
import re

text = """many SITUATIONS will CLEARLY Point to smaller models than larger ones, for example, real time response of less than 10 ms will often favour SLMs in edge devices than large models"""

# 1. Tokenization
tokens = word_tokenize(text)
print("Tokens:", tokens)

# 2. Lowercasing
lower_tokens = [token.lower() for token in tokens]
print("\nLowercased:", lower_tokens)

# 3. Alphabetic filtering
alpha_tokens = [token for token in lower_tokens if token.isalpha()]
print("\nAlpha only:", alpha_tokens)

# 4. Words > length 5
long_words = [w for w in alpha_tokens if len(w) > 5]
print("\nWords >5 chars:", long_words)  # ['situations', 'clearly', 'smaller', 'models', 'larger', 'example', 'response', 'often', 'favour', 'devices', 'large', 'models']

# 5. Lexical diversity (unique/total)
unique = len(set(alpha_tokens))
total = len(alpha_tokens)
diversity = unique / total
print(f"\nLexical diversity: {diversity:.3f}")
```

```python
import nltk
from nltk.corpus import brown
from nltk import ConditionalFreqDist, bigrams

nltk.download('brown')

# Get news category words
news_words = [w.lower() for w in brown.words(categories='news') if w.isalpha()]

# Create word → next-word conditional freq dist
cfd1 = ConditionalFreqDist(bigrams(news_words))
# Common words as conditions
common_words = [w for w, f in FreqDist(news_words).most_common(10)]

# Table: word → top 5 next words
cfd1.tabulate(conditions=common_words)
```
