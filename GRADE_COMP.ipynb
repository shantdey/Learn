```python
import pandas as pd
import numpy as np

# 1. Read data from the CSV dataset
df = pd.read_csv("Student_Grade.csv")

# Clean column names by stripping extra whitespaces/newlines
df.rename(columns=lambda x: x.strip(), inplace=True)

# Define the target column
col = "Total"

# Calculate the statistics
max_val = df[col].max()
min_val = df[col].min()
std_val = df[col].std() # Sample standard deviation

# 2. Develop logic for assigning grades
def assign_grade(val):
    if val == max_val:
        return "A+"
    elif val == min_val:
        return "D+"
    elif val >= max_val - 1 * std_val:
        return "A"
    elif val >= max_val - 1.5 * std_val:
        return "A-"
    elif val >= max_val - 2 * std_val:
        return "B+"
    elif val >= max_val - 2.5 * std_val:
        return "B"
    elif val >= max_val - 3 * std_val:
        return "B-"
    elif val >= max_val - 3.5 * std_val:
        return "C+"
    elif val >= max_val - 4 * std_val:
        return "C"
    else:
        # For values that fall below 4 standard deviations from the max but are not the minimum
        return "C-"

# Apply the logic to create a new "Grade" column
df["Grade"] = df[col].apply(assign_grade)

# Display the first few rows of the evaluated data
print(df.head(10))

# Save the evaluated data to a new CSV file
df.to_csv("Student_Grade_Evaluated.csv", index=False)
```
