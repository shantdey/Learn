#### Recursive functions
When a function called itself is called recursion. When we define a function and call the same function in their definition that means that code is recursive. If there is no terminating condition the recursion will get into infinite loop
* When a function called itself is called recursion.
* When we define a function and call the same function in their definition that means that code is recursive
* Classic example is creating a Fibonacci sequence or finding Factorial of a number

```python
# Program for fibonacci series up to k terms
def fib(k):
    # Base cases: return the value directly
    if k <= 1:
        return k
    else:
        # Recursive case: return the sum of previous two terms
        return fib(k-1) + fib(k-2)

n = int(input("value of k:"))

if n < 0:
    print("Incorrect input, please enter a non-negative integer")
else:
    print(f"Fibonacci sequence up to the {n}-th term:")
    for i in range(n):
        print(f"fib({i}): {fib(i)}")
    print(f"The {n}-th Fibonacci number is: {fib(n)}")

            
            
   
    

```

```python
# Recursive program for factorial of a number
def fact(k): # Recursive function
    if k == 1:
        return k
    else:
        return k * fact(k-1)
term = eval(input("Please enter a number: "))
if term <= 0:
    print ("Invalid input ! Please input a positive value")
elif term == 0:
    print ("Factorial of number 0 is 1")
else:
    print ("Factorial of number", term, "=", fact(term))
```

```python
# Alternate Implementation
def factorial(n):
    # Base case: stop recursion
    if n == 0 or n == 1:
        return 1
    # Recursive case: call itself
    else:
        return n * factorial(n - 1)

# Usage
result = factorial(5)
print(result)
```

### Binary Search
The binary search technique is to find search elements from a list of elements. 
* It works only for a sorted sequence of an element. 
* The first searching process starts by comparing the search element with the middle element in the list
* If the search element is smaller than the middle element, it will search the left side of the list and repeat it
* If the search element is greater than the middle element, it will search the right side of the list and repeat it
* The binary search process is implemented using recursion
* The worst-case time complexity is log n. It is efficient than a linear search algorithm

```python
# Program for iterative Binary Search Function
def binary_search(list2, x):
    beg = 0
    end = len(list2) - 1
    mid = 0
    while beg <= end:
        mid = (beg+ end) // 2
        if list2[mid] < x:
            beg = mid + 1
        elif list2[mid] > x:
            end = mid - 1
        else:return mid
    return -1 # Failure
# Test list
list1 = [2, 4, 3, 50, 40]
list1 = sorted(list1)
x = 40
# calling function
result = binary_search(list1, x)
if result != -1:
    print ("Element is present at index", str(result))
else:
    print ("Element is not present in list")
```

```python
# Program for recursive binary search
def binary_search(list2, low, high, x):
    # Check base case
    if high >= low:
        mid = (high + low) // 2
        if list2[mid] == x:
            return mid
        elif list2[mid] > x:
            return binary_search(list2, low, mid - 1, x)
        else:
            return binary_search(list2, mid + 1, high, x)
    else:
        return -1 # Not present in list
list1 = [20, 3, 4, 50, 40]
list1 = sorted(list1)
x = 20
# Function call
result = binary_search(list1, 0, len(list1)-1, x)
if result != -1:
    print ("Element is present at index", str(result))
else:
    print ("Element is not present in list")
```

#### Bubble sort
Bubble sort is a comparison-based algorithm that compares each pair of adjacent elements and swaps them if they are out of order. Since the average and worst-case complexity of this algorithm are O(n2), where n is the number of items, it is not suitable for large data sets.

```python
def bubble_sort(listBSort):
    for i in range(0,len(listBSort)-1):
        for j in range(len(listBSort)-1):
            if(listBSort[j]>listBSort[j+1]):
                temp = listBSort[j]
                listBSort[j] = listBSort[j+1]
                listBSort[j+1] = temp
    return listBSort
list1 = [5,3,8,7,6,2,1,4]
print("The sorted list is:", bubble_sort(list1))
```

#### Selection sort
* First elemeent of the list is chosen and compared with the remaining elements repeatedly
* elements are swapped if one is smaller than the other in ascending order
* Element in the second place in the list is then selected and same process is repeated
* Repeatedly check the lowest value from remainder part of the list

```python
def selection_sort(list1):
    for i in range(len(list1)):
        idx = i
        for j in range(i+1, len(list1)-1):
            if list1[j] < list1[idx]:
                idx = j
        list1[i], list1[idx] = list1[idx], list1[i]
listSSort = [32,23,27,44,35,25,18,7,2,0]
print(listSSort)
selection_sort(listSSort)
print(listSSort)
```

### Zip function
* Zip is used for combining the container of two different elements together
* The function takes iterables and combines their elements based on their order in the iterables
* As an input it takes any number of iterables
* Returns an iterator as tuples, input iterator can be list or strings

```python
L1 = [10,20,30,40,50]
L2 = ["S", "N", "O", "W", "Y"]
#combine using zip
result = tuple(zip(L1,L2))
print(result)
```

```python
#creating three lists of name, address and phone number and combining them
NameList = ["Alice", "Bob", "Charlie", "Donald"]
AddList = ["123 Street", "A22 Python Street", "Mall Road", "Donald Duck White House"]
PhoneNo = [12345, 67890, 111213, 141516, 171819]
result = tuple(zip(NameList, AddList, PhoneNo))
print(result)
for item in result:
    print("Name:", item[0])
    print("Address:", item[1])
    print('Phone No:', item[2])
    print("-"*20)
```

```python
# Inverse Zip function * with Zip
Tupa = [('Alice', 8), ("White Rabbit", 5), ('Chesire Cat', 10), ('Queen of Heart', 45)]
print(Tupa)
Tupa1, Tupa2 = zip(*Tupa)
print(Tupa1, Tupa2)
```

```python
# Function that returns multiple values
def readDate():
    """Get date from user and return as tuple."""
    print("Enter a date:")
    month = int(input("  month: "))
    day = int(input("  day: "))
    year = int(input("  year: "))
    return (month, day, year)  # Returns a tuple

# Method 1: Assign entire tuple to a variable
date = readDate()
print(date)  # Example Output: (5, 15, 2024)

# Method 2: Use tuple unpacking (more common)
month, day, year = readDate()
print(f"Date: {month}/{day}/{year}")  # Example Output: Date: 5/15/2024
```

```python
# Student Prepared
def fib(n):
    if n <= 0:
        print("input valid term")
    elif n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)

result = fib(10)
print(result)
```
