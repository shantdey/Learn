```python
import pypdf
import os

# 1. Open the PDF file in read-binary mode ('rb')
import pypdf

def extract_text_from_pdf(file_path):
    """
    Extracts text from all pages of a given PDF file.

    Args:
        file_path (str): The path to the PDF file.

    Returns:
        str: All extracted text combined into a single string.
    """
    full_text = ""  # Initialize an empty string to accumulate text

    # 1. Open the PDF file in binary read mode ('rb')
    with open(file_path, 'rb') as pdf_file:
        # 2. Create a PDF reader object
        pdf_reader = pypdf.PdfReader(pdf_file)

        # 3. Get the total number of pages
        num_pages = len(pdf_reader.pages)
        print(f"Total pages: {num_pages}")

        # 4. Loop through each page using the 'pages' property
        for page_num in range(num_pages):
            page = pdf_reader.pages[page_num]
            text = page.extract_text()
            if text: # Ensure text was actually extracted before appending
                full_text += text
                print(f"--- Page {page_num + 1} Content ---")
                print(text)

    # 5. Return all extracted text after the file is closed
    return full_text
```

### PyPDF Details
#### https://www.freecodecamp.org/news/how-to-work-with-pdf-files-in-python-a-pypdf-guide/
* The main class you’ll use is PdfReader for reading files and PdfWriter for creating or modifying them
* The first step in most tasks is opening a PDF file. PyPDF makes this simple using the PdfReader class
* Get the pages of a PDF file
* Each page in the document is represented as an object that you can access and work with.
* You can also inspect basic metadata such as the title or author if it’s available
* One of the most common use cases is text extraction. PyPDF allows you to extract text page by page
* It’s important to understand that text extraction isn’t perfect. PDFs store text based on layout, not reading order
* If you want to extract text from all pages, you can loop through them.
